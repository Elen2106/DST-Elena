```python
"""
    Авторами модуля представлена известная задача на поиск числа.
    Наиболее популярным является алгоритм Бинарного поиска.
    Для этого нам необходимо создать левую и правую границы загаданного числа.
    Алгоритм заключается в том, чтобы брать середину этого отрезка и сравнивать с ней число.
    Если число больше, чем середина, то сдвигаем левую границу.
    Если число меньше, чем середина, то сдвигаем правую границу.
    Иначе: Мы нашли загаданное число!
"""

import numpy as np

"""
    Предлагаю создать игру, в которой пользователь сам выбирает, 
    хочет ли загадать число компьютеру или наоборот угадывать.
    В конце каждой игры подводится итог - количество совершенных попыток.
"""


def search(x):  # Функция бинарного поиска, на выходе - количество попыток угадать число
    left = 0  # Левая граница поиска
    right = 10000  # Правая граница поиска
    count = 0  # Счетчик попыток

    while left < right - 1:  # Сам алгоритм поиска
        count += 1
        m = (left + right) // 2
        if x == m:
            break
        elif x > m:
            left = m
        elif x < m:
            right = m

    if count % 10 == 1 and count % 100 != 11:  # Подбор необходимого склонения
        return str(count) + ' попытку'
    elif 1 < count % 10 <= 4 and not (11 < count % 100 <= 14):
        return str(count) + ' попытки'
    else:
        return str(count) + ' попыток'


def yourSearch(searchNum):  # Функция, обрабатывающая попытки пользователя угадать число, на выходе - количество попыток
    x = int(input('Введите число:\n'))  # Первый ввод числа (возможно, что пользователь угадает с первого раза)
    count = 1  # Счетчик - первая попытка уже сделана

    while x != searchNum:  # Сам алгоритм поиска числа пользователем
        if x == searchNum:
            break
        elif x > searchNum:
            print(f'{x} больше, чем загаданное число')
        elif x < searchNum:
            print(f'{x} меньше, чем загаданное число')
        x = int(input('Введите число:\n'))
        count += 1

    if count % 10 == 1 and count % 100 != 11:  # Подбор необходимого склонения
        return str(count) + ' попытку'
    elif 1 < count % 10 <= 4 and not (11 < count % 100 <= 14):
        return str(count) + ' попытки'
    else:
        return str(count) + ' попыток'


def game():  # Главная функция игры, необходима для повторения главного меню,
    # в случае если пользователь введет посторонние символы
    mis = input('\nВыберите миссию:\n'
                '1. Вы вводите число, а компьютер пытается его угадать\n'
                '2. Компьютер загадывает число, а вы его угадываете\n'
                '3. Выход\n')

    if mis == '1':  # Переход к первой игре
        number1 = int(input('\nЗагадайте число компьютеру (до 10 000):\n'))
        print(f'Загаданное число {number1} было найдено за {search(number1)}')
        question = input('Вернуться в главное меню? (y/n):\n')
        if question == 'y':
            game()

    elif mis == '2':  # Переход ко второй игре
        print('Загадывает компьютер!')
        number2 = np.random.randint(1, 10000)
        print(f'Вы угадали загаданное компьютером число {number2} за {yourSearch(number2)}')
        print(f'А компьютер угадал бы это же число за {search(number2)}')
        question = input('\nВернуться в главное меню? (y/n):\n')
        if question == 'y':
            game()

    elif mis == '3':  # Выход из игры
        return

    else:  # Возврат в главное меню
        print('Попробуйте вписать другое число!')
        return game()


print('Поиграем в "Угадай число!"')  # Начало программы
game()
```
